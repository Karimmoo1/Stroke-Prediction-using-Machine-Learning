0. Feature Dictionary

```python
dataset_description = {
    "id": "Unique identifier",
    "gender": "Gender of the patient",
    "age": "Age of the patient",
    "hypertension": "Whether the patient has hypertension",
    "heart_disease": "Whether the patient has any heart diseases",
    "ever_married": "Marital status",
    "work_type": "Type of work",
    "Residence_type": "Urban or Rural",
    "avg_glucose_level": "Average glucose level in blood",
    "bmi": "Body Mass Index",
    "smoking_status": "Smoking status",
    "stroke": "Target variable (1 = stroke, 0 = no stroke)"
}

print("\n Dataset Feature Dictionary:")
for feature, description in dataset_description.items():
    print(f"{feature:20} → {description}")
```

1. Read & Preprocess 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_absolute_error, mean_squared_error, confusion_matrix, classification_report

from sklearn.linear_model import LogisticRegression, SGDClassifier, Perceptron
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.neural_network import MLPClassifier

from imblearn.over_sampling import SMOTE

# Load dataset
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
print(df.head())
print(df.shape)
print(df.info())
print(df.describe(include='all'))

# Drop ID column and fix missing
df.drop(['id'], axis=1, inplace=True)
df['bmi'] = df['bmi'].fillna(df['bmi'].median())

# Encode binary columns
binary_cols = ['gender', 'ever_married', 'Residence_type']
for col in binary_cols:
    df[col] = LabelEncoder().fit_transform(df[col])

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['work_type', 'smoking_status'], drop_first=True)

# Normalize numeric columns
scaler = StandardScaler()
df[['age', 'avg_glucose_level', 'bmi']] = scaler.fit_transform(df[['age', 'avg_glucose_level', 'bmi']])

# Split data
X = df.drop('stroke', axis=1)
y = df['stroke']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

2. Balance the Dataset 

```python
sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

# -------------------- 3. Visualization --------------------
plt.figure(figsize=(6, 4))
sns.countplot(x='stroke', data=df)
plt.title("Stroke Class Distribution (Original)")
plt.tight_layout()
plt.show()
```

4. Model Training

```python
models = {
    'LogisticRegression': LogisticRegression(),
    'GaussianNB': GaussianNB(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'SGDClassifier': SGDClassifier(),
    'LinearSVC': LinearSVC(),
    'Perceptron': Perceptron(),
    'MLPClassifier': MLPClassifier(max_iter=1000)  # Convergence warning suppressed
}

model_names = []
model_scores = []

for name, model in tqdm(models.items()):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    model_names.append(name)
    model_scores.append([
        accuracy_score(y_test, y_pred),
        precision_score(y_test, y_pred, zero_division=0),
        recall_score(y_test, y_pred, zero_division=0),
        f1_score(y_test, y_pred, zero_division=0),
        mean_absolute_error(y_test, y_pred),
        mean_squared_error(y_test, y_pred)
    ])
```

5. Metrics Table

```python
results_df = pd.DataFrame(model_scores, index=model_names, columns=['Accuracy', 'Precision', 'Recall', 'F1 Score', 'MAE', 'MSE'])
results_df = results_df.sort_values(by='F1 Score', ascending=False)
print("\n Model Evaluation Table:\n", results_df)
```

6. Best Model Evaluation

```python
best_model_name = results_df.index[0]
best_model = models[best_model_name]
y_pred_best = best_model.predict(X_test)

print(f"\n Best Model: {best_model_name}")
print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
print("\n Classification Report:\n", classification_report(y_test, y_pred_best, zero_division=0))

plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred_best), annot=True, fmt='d', cmap='Blues')
plt.title(f"Confusion Matrix - {best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

7. Visualization of Metrics

```python
results_df[['Accuracy', 'Precision', 'Recall', 'F1 Score']].plot(kind='bar', figsize=(12, 6))
plt.title('Model Performance on Stroke Prediction (After SMOTE)')
plt.xlabel('Model')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
